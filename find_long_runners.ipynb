```python
import pandas as pd
```

```python
apexdf = pd.read_csv('~/2021-01-04_ApexExecution.csv')
```

```python
#apexdf
long_running = apexdf[apexdf['RUN_TIME']>5000].copy(deep=True)

quids = ['E','H','I','K','L','M','R','V']
long_running = long_running[long_running['QUIDDITY'].isin(quids)]
long_running
```

```python
import math
run_diff = long_running['TIMESTAMP'].diff().to_list()
def nm(x):
    if (math.isnan(x)) : return 0
    return x
run_diff  = [math.floor(nm(x) * 1000) for x in run_diff]
long_running['RUN_DIFF'] = run_diff
```

```python
long_running[['TIMESTAMP_DERIVED', 'ENTRY_POINT', 'RUN_TIME', 'RUN_DIFF', 'QUIDDITY' ]]

```

```python
run_time = long_running['RUN_TIME'].to_list()
run_diff = long_running['RUN_DIFF'].to_list()

overlap = [run_diff[c+1] - run_time[c] for c in range(len(run_time)-1)]
overlap.insert(0,0)

long_running['OVERLAP'] = overlap
```

```python
myrows = long_running[['TIMESTAMP_DERIVED', 'ENTRY_POINT', 'RUN_TIME', 'RUN_DIFF', 'OVERLAP', 'QUIDDITY' ]]
```

```python
myrows.to_csv('experted.csv')
```
