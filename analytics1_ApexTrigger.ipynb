```python
import pandas as pd
```

```python
df = pd.read_csv('2020-08-09_ApexTrigger.csv')
```

```python
df
```

# Trigger run stats

```python
agg = df.groupby('TRIGGER_NAME').agg({
    'CPU_TIME' : ['count', 'mean'],
    'EXEC_TIME' : ['mean']
})
combi_cols = [col + ':' + fun for (col, fun) in  agg.columns]
agg.columns = combi_cols
```

```python
agg.sort_values(by='CPU_TIME:count', ascending=False)
```

# User run stats

```python
user_agg = df.groupby('USER_ID').agg({
    'CPU_TIME' : ['count', 'mean'],
    'EXEC_TIME' : ['mean']
})
combi_cols = [col + ':' + fun for (col, fun) in  user_agg.columns]
user_agg.columns = combi_cols
```

```python
user_agg.sort_values(by='CPU_TIME:count', ascending = False)[0:30]
```
